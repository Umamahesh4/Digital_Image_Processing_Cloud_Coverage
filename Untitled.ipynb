```python
import cv2
import numpy as np

def calculate_rms(original, transformed):
    
    original = cv2.resize(original, (transformed.shape[1], transformed.shape[0]))

    return np.sqrt(np.mean((original.astype(np.float32) - transformed.astype(np.float32)) ** 2))

def resize_image(image_path, output_path_nn, output_path_bl, output_path_bc, new_size=(256, 256)):
    # Load the image
    image = cv2.imread(image_path)
    
    if image is None:
        print(f"Error loading image: {image_path}")
        return

    # Nearest Neighbor Interpolation
    resized_nn = cv2.resize(image, new_size, interpolation=cv2.INTER_NEAREST)
    cv2.imwrite(output_path_nn, resized_nn)

    # Bilinear Interpolations
    resized_bl = cv2.resize(image, new_size, interpolation=cv2.INTER_LINEAR)
    cv2.imwrite(output_path_bl, resized_bl)

    # Bicubic Interpolation
    resized_bc = cv2.resize(image, new_size, interpolation=cv2.INTER_CUBIC)
    cv2.imwrite(output_path_bc, resized_bc)

    # Calculate RMS Errors
    rms_nn = calculate_rms(image, resized_nn)
    rms_bl = calculate_rms(image, resized_bl)
    rms_bc = calculate_rms(image, resized_bc)

    print(f'Resized image saved to {output_path_nn}')
    print(f'RMS Error (Nearest Neighbor): {rms_nn:.2f}')
    print(f'Resized image saved to {output_path_bl}')
    print(f'RMS Error (Bilinear): {rms_bl:.2f}')
    print(f'Resized image saved to {output_path_bc}')
    print(f'RMS Error (Bicubic): {rms_bc:.2f}')

# Example usage
if __name__ == "__main__":
    for i in range(1, 301):
        input_image_path = f'cloudy_new{i}.jpg'  # Replace with your image path
        output_image_path_nn = f'near_neighbour/cloudy_nn{i}.jpg'  # Output path for nearest neighbor resized image
        output_image_path_bl = f'bilinear/cloudy_bl{i}.jpg'  # Output path for bilinear resized image
        output_image_path_bc = f'bicubic/cloudy_bc{i}.jpg'  # Output path for bicubic resized image
        
        resize_image(input_image_path, output_image_path_nn, output_image_path_bl, output_image_path_bc)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calculate_rms(original, transformed):
    """Calculate the Root Mean Square error between original and transformed images."""
    # Resize the original image to match the transformed image size
    original_resized = cv2.resize(original, (transformed.shape[1], transformed.shape[0]))
    error = np.mean((original_resized.astype(np.float32) - transformed.astype(np.float32)) ** 2)
    return np.sqrt(error)

def resize_and_calculate_rms(image_path, new_size=(256, 522)):
    """Resize an image using different methods and calculate RMS error."""
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading image: {image_path}")
        return None, None, None

    # Resizing images
    resized_nn = cv2.resize(image, new_size, interpolation=cv2.INTER_NEAREST)
    resized_bl = cv2.resize(image, new_size, interpolation=cv2.INTER_LINEAR)
    resized_bc = cv2.resize(image, new_size, interpolation=cv2.INTER_CUBIC)

    # Calculating RMS error for each method
    rms_nn = calculate_rms(image, resized_nn)
    rms_bl = calculate_rms(image, resized_bl)
    rms_bc = calculate_rms(image, resized_bc)

    return rms_nn, rms_bl, rms_bc

def plot_rms_values(rms_values, labels=['Nearest Neighbor', 'Bilinear', 'Bicubic']):
    """Plot RMS values for different interpolation methods."""
    indices = np.arange(len(rms_values[0]))
    plt.figure(figsize=(12, 6))
    for i, rms in enumerate(rms_values):
        plt.plot(indices, rms, label=labels[i], marker='o')
    
    plt.xlabel('Image Index')
    plt.ylabel('RMS Error')
    plt.title('RMS Error Comparison of Interpolation Methods')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    rms_nn_list, rms_bl_list, rms_bc_list = [], [], []
    for i in range(1, 301):  # Assuming 300 images
        image_path = f'cloudy_new{i}.jpg'
        rms_results = resize_and_calculate_rms(image_path)
        if rms_results:
            rms_nn_list.append(rms_results[0])
            rms_bl_list.append(rms_results[1])
            rms_bc_list.append(rms_results[2])
    
    # Plot the RMS values
    plot_rms_values([rms_nn_list, rms_bl_list, rms_bc_list])
```
